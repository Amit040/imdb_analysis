```python
# Divide the 'gross' and 'budget' columns by 1000000 to convert '$' to 'million $'
movie['budget']=movie['budget']/1000000
movie['Gross']=movie['Gross']/1000000
movie
```

```python
#change the sacle of matrix
movie['MetaCritic']= movie['MetaCritic']/10
movie
```

```python
# Find the average ratings
movie['Avg_rating']= (movie["IMDb_rating"]+movie["MetaCritic"])/2
movie
```

```python
#Sort in descending order of average rating
df=movie[["Title","IMDb_rating","MetaCritic","Avg_rating"]]
df= df.loc[abs(df["IMDb_rating"]-df['MetaCritic']<0.5)]
df
```

```python
df = df.sort_values(by = 'Avg_rating',ascending = False)
```
